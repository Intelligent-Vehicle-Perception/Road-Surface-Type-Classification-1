<h5>Models</h5>

```
def model1(n_convolutional, n_features, n_outputs):
    model = Sequential()
    model.add(Conv1D(filters=64, kernel_size=3, activation='relu', input_shape=(n_convolutional, n_features)))
    model.add(Conv1D(filters=64, kernel_size=3, activation='relu'))
    model.add(Conv1D(filters=128, kernel_size=3, activation='relu'))
    model.add(Flatten())
    model.add(Dense(100, activation='relu'))
    model.add(Dense(n_outputs, activation='softmax'))
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['acc'])
    return model, "CNN/CNN 1"

def model2(n_convolutional, n_features, n_outputs):
    model = Sequential()
    model.add(Conv1D(filters=64, kernel_size=3, activation='relu', input_shape=(n_convolutional, n_features)))
    model.add(Conv1D(filters=64, kernel_size=3, activation='relu'))
    model.add(Conv1D(filters=128, kernel_size=3, activation='relu'))
    model.add(MaxPooling1D(pool_size=2))
    model.add(Flatten())
    model.add(Dense(100, activation='relu'))
    model.add(Dense(n_outputs, activation='softmax'))
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['acc'])
    return model, "CNN/CNN 2"

def model3(n_convolutional, n_features, n_outputs):
    model = Sequential()
    model.add(Conv1D(filters=64, kernel_size=3, activation='relu', input_shape=(n_convolutional, n_features)))
    model.add(BatchNormalization())
    model.add(Dropout(0.15))
    model.add(Conv1D(filters=32, kernel_size=3, activation='relu'))
    model.add(GlobalAvgPool1D())
    model.add(BatchNormalization())
    model.add(Dropout(0.2))
    model.add(Dense(n_outputs, activation='softmax'))
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['acc'])
    return model, "CNN/CNN 3"

# https://www.sciencedirect.com/science/article/pii/S0966636219307064
def model4(n_convolutional, n_features, n_outputs):
    model = Sequential()
    model.add(Conv1D(filters=100, kernel_size=10, activation='relu', input_shape=(n_convolutional, n_features)))
    model.add(Conv1D(filters=100, kernel_size=10, activation='relu'))
    model.add(MaxPooling1D(pool_size=3))
    model.add(Conv1D(filters=160, kernel_size=10, activation='relu'))
    model.add(Conv1D(filters=160, kernel_size=10, activation='relu'))
    model.add(GlobalAvgPool1D())
    model.add(Dropout(0.5))
    model.add(Dense(n_outputs, activation='softmax'))
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['acc'])
    return model, "CNN/CNN 4"

# https://ieeexplore.ieee.org/document/8559719
def model5(n_convolutional, n_features, n_outputs):
    model = Sequential()
    model.add(Conv1D(filters=64, kernel_size=3, activation='relu', input_shape=(n_convolutional, n_features)))
    model.add(MaxPooling1D(pool_size=2))
    model.add(Conv1D(filters=64, kernel_size=3, activation='relu'))
    model.add(Flatten())
    model.add(Dense(n_outputs, activation='softmax'))
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['acc'])
    return model, "CNN/CNN 5"

# https://blogs.rstudio.com/tensorflow/posts/2018-07-17-activity-detection/
def model6(n_convolutional, n_features, n_outputs):
    model = Sequential()
    model.add(Conv1D(filters=24, kernel_size=8, activation='relu', input_shape=(n_convolutional, n_features)))
    model.add(BatchNormalization())
    model.add(SpatialDropout1D(0.15))
    model.add(Conv1D(filters=12, kernel_size=8, activation='relu'))
    model.add(GlobalAvgPool1D())
    model.add(BatchNormalization())
    model.add(Dropout(0.2))
    model.add(Dense(48, activation='relu'))
    model.add(BatchNormalization())
    model.add(Dropout(0.25))
    model.add(Dense(n_outputs, activation='softmax'))
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['acc'])
    return model, "CNN/CNN 6"

# https://ieeexplore.ieee.org/document/8834480
def model7(n_convolutional, n_features, n_outputs):
    model = Sequential()
    model.add(Conv1D(filters=128, kernel_size=3, activation='relu', input_shape=(n_convolutional, n_features)))
    model.add(BatchNormalization())
    model.add(MaxPooling1D(pool_size=2))
    model.add(Conv1D(filters=64, kernel_size=3, activation='relu'))
    model.add(BatchNormalization())
    model.add(MaxPooling1D(pool_size=2))
    model.add(Conv1D(filters=32, kernel_size=3, activation='relu'))
    model.add(BatchNormalization())
    model.add(MaxPooling1D(pool_size=2))
    model.add(Flatten())
    model.add(Dropout(0.2))
    model.add(Dense(n_outputs, activation='softmax'))
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['acc'])
    return model, "CNN/CNN 7"
```

```
def bestModel(n_convolutional, n_features, n_outputs):
    model = Sequential()
    model.add(Conv1D(filters=64, kernel_size=3, activation='relu', input_shape=(n_convolutional, n_features)))
    model.add(BatchNormalization())
    model.add(Dropout(0.15))
    model.add(Conv1D(filters=32, kernel_size=3, activation='relu'))
    model.add(BatchNormalization())
    model.add(Dropout(0.15))
    model.add(Conv1D(filters=32, kernel_size=3, activation='relu'))
    model.add(GlobalAvgPool1D())
    model.add(BatchNormalization())
    model.add(Dropout(0.2))
    model.add(Dense(32, activation='relu'))
    model.add(Dense(n_outputs, activation='softmax'))
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['acc'])
    return model, "CNN/CNN Best Model"
```
