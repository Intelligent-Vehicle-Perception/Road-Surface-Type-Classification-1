<h5>Execution Logs</h5>

```
def getExecutionLogs(dnnType):

    execution_logs = {}
    columns = ['window', 'train_loss',	'val_loss',	'train_acc', 'val_acc']
    models = os.listdir(os.path.join(work_folder, dnnType))

    for model in models:
        
        appended = []
        
        for experiment in range(1, 4):

            file = os.path.join(work_folder, dnnType, model, "Experiment " + str(experiment), "experiment-execution-log.csv")

            if os.path.isfile(file):
                content = pd.read_csv(file)
                appended.append(content[columns])

        execution_logs[model] = appended    
    
    return execution_logs
```

```
def printValuesDnn(dnnType, include=None, exclude=None, statistics=False):

    execution_logs = getExecutionLogs(dnnType)

    for model in execution_logs.keys():
        
        experiments = len(execution_logs[model])
        sum = None
        
        for i in range(0, experiments):

            if sum is None:
                sum = execution_logs[model][i].copy()
            else:
                sum += execution_logs[model][i].copy()
        
        if experiments > 0:

            if include is not None:

                show = False

                for part in include:
                    if part in model:
                        show=True
                        break

            elif exclude is not None:

                show = True

                for part in exclude:
                    if part in model:
                        show=False
                        break

            else:
                show = True

            if show:
                pd.set_option("float_format", '{:0.5f}'.format)
                print("Model " + model + (" | Experiments " + str(experiments) if experiments < 3 else ""))
                result = sum.copy()/experiments
                display(result)

                if statistics:
                    pd.set_option("float_format", '{:0.2f}%'.format)
                    result = result[['window','val_acc']]
                    result = result.transpose()
                    result["average"] = result.mean(axis=1) 
                    display(result * 100)

                print("\n")
```

```
printValuesDnn('LSTM', exclude=["Test","Best"], statistics=True)
```
